```python
# Seleccionamos las columnas entre StartDate y LastUpdatePosted
df.loc[:,'StartDate':'LastUpdatePosted']
```

```python
# Observamos que en la columna  StartDate todas las observaciones
# no siguen el mismo formato "%B %d, %Y"
# Convenio : En las observaciones (columna StartDate) que non sigan el
# formato consideraremos que la fecha proviene del 1 de cada mes

# Trabajo 1 : Reconocer las filas que no cuentan con la informacion del dia
ListaIndices = list()
for indice in range(df.shape[0]):
  # Si no tiene la coma que indica el dia
  # si no es nan
  if (str(df.StartDate[indice]).rfind(",") == -1) and (df.isnull().StartDate[indice] == False) :
    ListaIndices.append(indice)

# Ya encontramos la lista de indices que hay que modificar :ListaIndices
# Regla de contruccion para agregar el 1 de cada mes a la fecha
for indice in ListaIndices:
  # print(df.StartDate[indice])
  lista_temp = str(df.StartDate[indice]).split(" ")
  lista_temp.insert(-1, "1,")
  df.StartDate[indice] = " ".join(elemento for elemento in lista_temp)

for indice in ListaIndices:
  print(df.StartDate[indice])
```

```python
# Necesitamos recuperar lo que se encuentra entre parentesis de cada
# observacion .
# Dada la naturaleza del problema, debemos usar expresiones ragulares
import re
```

```python
# Listemos el contenido del modulo re
dir(re)
```

```python
# Veamos la documentacion de algunas de las funciones del modulo re
help(re.findall)
```

```python
re.findall(r'\(.*\)', df.Age[5000])
```

```python
# Empaquetemos este filtro en una funcion
def LimpiezaAge(edad):
  # La convencio es que recuperamos lo que se encuentra entre parentesis
  # Pero el codigo tambien debe estar preparado para el caso
  # en que no se cumple lo que habiamos pensado . Puede haber observaciones
  # que no tengan la informacion entre parentesis
  if len(re.findall(r'\(.*\)', edad)):
    return re.findall(r'\(.*\)', edad)[0]
  # Si en una observacion no se encuentran los parentesis, no modifica
  # la observacion
  return edad
```

```python
# Prueba 1  de LimpiezaAge
LimpiezaAge(df["Age"][2024])
```

```python
# Aplicar la funcion LimpiezaAge a cada observacion de la
# columna Age
df["Age"] = df["Age"].apply(lambda x: LimpiezaAge(x))
```
