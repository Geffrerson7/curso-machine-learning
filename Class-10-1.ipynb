```python
import os
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
#skearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn import metrics
#skearn : Modelos
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import tree
from sklearn.datasets import load_iris
```

```python
#datasets
iris = load_iris()
# Particionamiento de los datos
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target,test_size = 0.2)
```

```python
#Definamos 3 pipelines
pipe_reglog = Pipeline([
    ("scl",StandardScaler()),
    ("reglog_clf", LogisticRegression(max_iter=1000,fit_intercept=False))
])

pipe_svm = Pipeline([
    ("scl", StandardScaler()),
    ("pca", PCA(n_components = 2)),
    ("svm_clf", svm.SVC())
])

pipe_dt = Pipeline([
    ("scl", StandardScaler()),
    ("pca", PCA(n_components = 3)),
    ("dt_clf", tree.DecisionTreeClassifier())
])
```

```python
#Lista de pipelines
pipelines = [pipe_reglog, pipe_svm, pipe_dt]
#Definamos un diccionario de pipelines y clasificadores para referenciar
pipe_dict ={0:"Regresion Logistica", 1:"Maquina de Soporte Vectorial", 2:"Arbol de decision"}

# "Ajustar" los pipeline
for p in pipelines:
    p.fit(x_train, y_train)
    
```

```python
# Comparemos accuracies
for idx, val in enumerate(pipelines):
  print("%s Pipeline - Score (Test) : %f" %(pipe_dict[idx], val.score(x_test, y_test)))
```

```python
# Calculemos indicadores de calidad (test)

for idx, val in enumerate(pipelines):
  y_forecast = val.predict(x_test)
  kpi = metrics.accuracy_score(y_test, y_forecast)
  print("Modelo : %s  - Acc : %f" %(pipe_dict[idx], kpi))
```
