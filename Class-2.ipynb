```python
# Implementemos el producto interno de dos vectores generados de
# manera aleatoria usando listas . El Numero de elementos de los
# vectores es ingresado por el usuario como unico argumento.
# Lo que la funcion devuelve es el producto de estos vectores
#
# Actualizacion : Se requiere que la funcion devuelva tambien los
# dos vectores generados de manera aleatoria
#
# Segunda Actualizacion : El output debe estar empaquetado en
# un diccionario

def prodInterno(n):
  vec1 = []
  vec2 = []

  # Creacion de los elementos de los vectores
  for e in range(n):
    vec1.append(rnd.random())
    vec2.append(rnd.random())

  # Calculo del producto interno
  pi = 0
  for e in range(n):
    pi = pi + vec1[e]*vec2[e]

  # revolvamos la variable pi
    # Agregamos el nuevo requerimiento
      # Agreguemos el segundo requerimiento
  return {"ProdInterno": pi, "vec1": vec1, "vec2":vec2}
```
