```python
# No vamos a considerar la informacion GMT (puesto que se repite en todas
# las observaciones)

Pref_comida["Timestamp"].str[:-6]
```

```python
# Formateamos Timestamp
Pref_comida["Timestamp1"] = pd.to_datetime(Pref_comida["Timestamp"].str[:-6],
                                          #  Espefificamos el formato que tiene la informacion
                                           format = "%Y/%m/%d %H:%M:%S %p")

```

```python
# Metodo dtypes
Pref_comida.dtypes
```

```python
# Separamos la informacion de año/mes/dia
Pref_comida["TimestampFecha"] = Pref_comida["Timestamp1"].dt.strftime("%Y/%m/%d")

# Separemos la informacion de hora:minuto:segundo
Pref_comida["TimestampHora"] = Pref_comida["Timestamp1"].dt.strftime("%H:%M:%S")

# Informacion AM/PM
Pref_comida["TimestampAMPM"]  = Pref_comida["Timestamp1"].dt.strftime("%p")

# Cambiemos de nombre a Timestamp
Pref_comida.rename(columns = {"Timestamp": "Date"}, inplace= True)

# Definamos a la columna Date como el conjunto de indices
Pref_comida.set_index("Date", inplace = True)
```

```python
# Veamos la informacion que tenemos en la columna nacionalidad
# Listemos los valores que puede tomar la columna Nationality
Pref_comida["Nationality"].unique()
```

```python
# Como podemos observar en la celda anterior, las nacionalidades estan escritas
# o bien mal o bien con diferente formato .Ademas Muslim (musulman) no es una
# nacionalidad, es una religion

# Correcciones
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["Pakistani", "Pakistan", "Pakistani"], "pakistani")
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["Indonesia", "Indonesian ", "Indonesian", "Indonesain"], "indonesian")
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["Muslim"], "Unknown")
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["Maldivian", "Maldivian "],"maldivian" )
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["MY", "Malaysian","Malaysian ", "MALAYSIAN", "Malaysia ","Malaysia"], "malaysian")
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["Japan"], "japanese")
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["China"], "chinese")
Pref_comida["Nationality"] = Pref_comida["Nationality"].replace(["Yemen"], "yemeni")
```

```python
# Verifiquemos las nacionalidades
Pref_comida.Nationality.unique()
```

## Analicemos estadisticamente los datos

```python
# Construyamos una copia de los datos originales
data = Pref_comida.copy()
data.columns
```

```python
# Observemos en un diagrama de barras la variable Nationality
sns.displot(data["Nationality"])

# Afortunadamente seaborn esta construido sobre matplotlib
plt.xticks(rotation = 90)
plt.ylabel("Frecuencia")
plt.xlabel("Nacionalidades")
plt.show()
```

```python
# Como la diferencia en el numero de observaciones de nacionalidad India es bastante
# con respecto a las otras nacionalidades . Consideremos todas las nacionalidades
# excepto a la India
sns.displot(data[data.Nationality != "Indian"]["Nationality"])

# Afortunadamente seaborn esta construido sobre matplotlib
plt.xticks(rotation = 90)
plt.ylabel("Frecuencia")
plt.xlabel("Nacionalidades")
plt.show()
```

```python
# Ahora veamos la preferencia en el tipo de comida (columna Food) de las personas
# de la India

indianFoodChoice = data[data.Nationality == "Indian"]["Food"].value_counts()

# Mostremos esta informacion usando la API de matplotlib en pandas
indianFoodChoice.plot(kind = "barh", color = "#5A7A43")

# Modifiquemos las etiquetas
plt.xlabel("Frecuencia")
plt.ylabel("Tipo de comida")
plt.title("Preferencias de las personas de la India")
plt.show()
```

```python
# Desagreguemos el diagrama de barras anterior considerando la variable Genero

# Las personas de nacionalidad india
indians = data[data.Nationality == "Indian"]
indians = indians.groupby("Gender")["Food"].value_counts().unstack()
indians.plot(kind = "barh")

# modifiquemos las etiquetas
plt.xlabel("Frecuencia")
plt.title("Preferencia de comida por genero : Indian")
plt.show()
```

```python
# Preferencias de los postres : Dessert
Ind_dessertPref = data[data.Nationality == "Indian"]
sns.displot(Ind_dessertPref["Dessert"])
```

```python
# Age : Veamos la distribucion de edades mediante un histograma
data.Age.plot(kind = 'hist')
plt.xlabel("Edad")
plt.title("Numero de Obs por edad")
plt.show()
```

```python
# Consideremos la poblacion cuyas edades estan en [20,30]
# para observar las preferencias del tipo de jugo : Juice
pref_juice = data[(data.Age >=20) & (data.Age <= 30)]["Juice"].value_counts()
pref_juice.plot(kind = "bar")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Podemos desagregar el grafico anterior considerando el genero
genderPref = data[(data.Age >=20) & (data.Age <= 30)].groupby("Gender")["Juice"].value_counts().unstack()
genderPref.plot(kind = "bar")
plt.ylabel("Frecuencia")
plt.xlabel("Genero")
plt.title("Gender vs Juice")
```

## Codificando caracteristicas (variables)

```python
# NIveles de la variables Dessert
data["Dessert"].unique()
# 'Maybe', 'Yes', 'No'

# Codifiquemos la columnas Dessert: traducir la informacion cualitativa a
# informacion cuantitativa
dic = {"Yes":2, "Maybe":1, "No":0}
dic1 = {"Yes":1, "Maybe":0, "No":-1}

# Crear una nueva variables para tener una version codificada de Dessert
data["Rating_Dessert"] = data.Dessert.map(dic)
data["Rating_Dessert1"] = data.Dessert.map(dic1)

```

```python
# Analicemos a Rating_Dessert
data.Rating_Dessert.unique()
```

```python
# Analicemos a Rating_Dessert1
data.Rating_Dessert1.unique()
```

```python
# Sabemos que Gender, Food, Juice son variables binarias
# Utilicemos otra estrategia para codificar estas variables
data["Gender_bin"] = data["Gender"].apply(lambda x : 1 if x == "Male" else (0 if x=="Female" else None))
# data["Gender_bin"].unique()

data["Food_bin"] = data["Food"].apply(lambda x : 1 if x == "Traditional food" else (0 if x =="Western Food" else None))

data["Juice_bin"] = data["Juice"].apply(lambda x: 1 if x=="Fresh Juice" else (0 if x=="Carbonated drinks" else None) )

```
